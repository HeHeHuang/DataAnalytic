###  Be Careful when copying arrays!!!


```python
#
a = np.array([1,2,3])
b = a # not make a new copy but make two tag point to same thing
b[0]=100 
print(a)
print(b)
```

```python
#
c = np.array([1,2,3])
d= c.copy()# not make a new copy but make two tag point to same thing
d[0]=100 
print(d)
print(c)
```
